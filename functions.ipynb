```python
def cal_fact(n) :
  if(n == 1 or  n == 0):
    return 1
  return cal_fact(n-1) * n
cal_fact(6)
```

```python
def natural (n) :
  if(n == 0):
    return 0
  return natural(n-1) + n


sum = natural(5)
print(sum)
```
