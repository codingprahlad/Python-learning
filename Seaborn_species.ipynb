```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
spec = sns.load_dataset("iris")
```

```python
print(spec)
```

```python
print(spec.head())
```

```python
spec.species.unique()
```

```python
spec.species.value_counts()
```

```python
sns.lineplot(x="sepal_length", y = "sepal_width", data=spec, ci=None)
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
sns.lineplot(x="sepal_length", y = "sepal_width", data=spec, ci=None)
plt.show()
```

```python
sns.lineplot(x='sepal_length', y='species', data=spec, hue='species')
plt.show()
```

```python
sns.scatterplot(x='petal_length',y='sepal_length',data=spec,hue='species')
plt.title("Scatter plot")
plt.show()
```

```python
sns.barplot(x = "species", y ="sepal_length", data = spec)
plt.show()
```

```python
sns.countplot(x ="species", data=spec)
plt.show()
```

```python
sns.boxplot(x="sepal_width", data=spec, hue="species")
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_width", data=spec)
plt.show()
```

```python
sns.violinplot(x="species", y = "sepal_width", data=spec)
plt.show()
```

```python
sns.stripplot(x="species", y = "sepal_width", data=spec)
plt.show()
```

```python
sns.histplot(x="petal_width", data=spec, kde=True)
plt.show()
```

```python
sns.histplot(x="sepal_width", data= spec, kde=True)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.displot(spec['petal_width'],color="g", kde=True) # ,hist=False, kde=False
plt.show()
```

```python
sns.pairplot(data=spec, hue='species')
plt.show()
```

```python
sns.catplot(x='sepal_length', data=spec, kind='violin')
plt.show()
```

```python
sns.catplot(x='petal_length',y='species',data=spec,kind='swarm')
plt.show()
```

```python
tc = spec.drop("species",axis=1).corr()
```
